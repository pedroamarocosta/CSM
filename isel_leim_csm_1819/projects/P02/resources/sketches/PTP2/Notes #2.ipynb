```python
from time import time
from os import path
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
def generateHuffman():
    
    symbolData = np.fromfile('testHuffman.txt', 'uint8')
    #symbolData = np.fromfile('iliad.txt', 'uint8')
    
    print("File size = " + str(os.path.getsize('iliad.txt')/1000.) + " Kb" + "\n")
    
    symbolSet = np.unique(symbolData)
    
    symbolReps = np.array([np.sum(symbolData==i) for i in symbolSet])
    
    indexSorted = np.argsort(symbolReps)
    
    symbolSet = symbolSet[indexSorted]
    
    charSet = [chr(i) for i in symbolSet]
    
    symbolTableBinary = [np.empty(0, dtype=np.uint8) for i in range(len(symbolSet))]
    
    symbolTableString = ['' for i in range(len(symbolSet))]
  
    symbolIndexes = [(i,) for i in range(len(symbolReps))]
    
    symbolReps = symbolReps[indexSorted]
    
    symbolTuple = [(i,) for i in symbolSet]
    symbolTuple.append("EOF")
    symbolTuple = np.delete(symbolTuple, len(symbolTuple)-1, None)
    
    print("indexSorted = " + str(indexSorted) + "\n")
    print("symbolIndexes = " + str(symbolIndexes) + "\n")
    print("symbolTuple = " + str(symbolTuple) + "\n")
    print("symbolReps = " + str(symbolReps) + "\n")
    print("symbolTableBinary = " + str(symbolTableBinary) + "\n")
    print("symbolTableString = " + str(symbolTableString) + "\n")
    print("______________________________________________________________________________________________")
    print("\n")
    
    while(len(symbolReps) > 1):
       
        i0 = 0
        i1 = 1
        
        workindIndexes = np.array(symbolIndexes[i0] + symbolIndexes[i1])
        
        c0 = workindIndexes[:int(len(workindIndexes)/2)]
        c1 = workindIndexes[int(len(workindIndexes)/2):]
        
        symbolIndexes[i1] = symbolIndexes[i0] + symbolIndexes[i1]
        symbolIndexes = np.delete(symbolIndexes, i0)
        
        symbolReps[i1] = symbolReps[i0] + symbolReps[i1]
        symbolReps = np.delete(symbolReps, i0)
        
        symbolTuple[i1] = symbolTuple[i0] + symbolTuple[i1]
        symbolTuple = np.delete(symbolTuple, i0)
        
        indexSorted = np.argsort(symbolReps)
        
        symbolReps = symbolReps[indexSorted]
        symbolIndexes = symbolIndexes[indexSorted]
        symbolTuple = symbolTuple[indexSorted]     
        
        for j in range(len(c0)):
            symbolTableString[c0[j]] = "0" + symbolTableString[c0[j]]
        
        for k in range(len(c1)):
            symbolTableString[c1[k]] = "1" + symbolTableString[c1[k]]
        
        
        for j in range(len(c0)):
            symbolTableBinary[c0[j]] = np.insert(symbolTableBinary[c0[j]], 0, 0)
        
        for k in range(len(c1)):
            symbolTableBinary[c1[k]] = np.insert(symbolTableBinary[c1[k]], 0, 1)
        
        print("workindIndexes = " + str(workindIndexes) + "\n")
        print("c0 = " + str(c0) + "\n")
        print("c1 = " + str(c1) + "\n")
        print("indexSorted = " + str(indexSorted) + "\n")
        print("symbolIndexes = " + str(symbolIndexes) + "\n")
        print("symbolTuple = " + str(symbolTuple) + "\n")
        print("symbolReps = " + str(symbolReps) + "\n")
        print("symbolTableString = " + str(symbolTableString) + "\n")
        print("symbolTableBinary = " + str(symbolTableBinary) + "\n")
        print("______________________________________________________________________________________________")
        print("\n")

    print("symbolData = " + str(symbolData) + "\n")
    print("symbolSet = " + str(symbolSet) + "\n")
    print("symbolIndexes = " + str(symbolIndexes) + "\n")
    print("symbolReps = " + str(symbolReps) + "\n")
    print("indexSorted = " + str(indexSorted) + "\n")
    print("symbolTuple = " + str(symbolTuple) + "\n")
    print("symbolTableString = " + str(symbolTableString) + "\n")
    print("symbolTableBinary = " + str(symbolTableBinary) + "\n")
    
    return symbolSet, symbolReps, charSet, symbolTableBinary, symbolTableString

s, r, c, tb, ts = generateHuffman()

print("Huffman Table (binary array)" + "\n")
for i in range(len(tb)):
    print(str(c[i]) + " " + str(tb[i])) 
print("\n")

print("Huffman Table (string array)" + "\n")
for i in range(len(ts)):
    print(str(c[i]) + " " + str(ts[i])) 
print("\n")
```
