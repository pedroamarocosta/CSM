<h5>2. Grave a mesma imagem, mas agora em formato "JPEG" com diferentes qualidades. Verifique visualmente a qualidade das imagens assim como o tamanho do ficheiro. Calcule a taxa de compress√£o, a SNR e a PSNR.</h5>
<hr>

```python
def calculateCR(original_file, compressed_file):

    return os.stat(original_file).st_size/os.stat(compressed_file).st_size

def calculateSNR(original_image, compressed_image):
    
    signal = np.sum(np.power(compressed_image, 2))
    error = np.sum(np.power((original_image - compressed_image), 2))
    return 10*np.log10(signal/error)


def calculatePSNR(original_image, compressed_image):
    
    max_value = np.power(np.max(original_image), 2)
    mean_squared_error = (1/(3*compressed_image.shape[0]*compressed_image.shape[1])) * \
                        np.sum(np.power((original_image-compressed_image), 2))
    return 10 * np.log10(max_value/mean_squared_error)


def compress_to_jpg(img_loc, quality):
    
    save_name = 'ex2_tiff_to_jpg/img_quality_'+ str(quality) + '.jpg'
    
    original_img = cv2.imread(img_loc)
    
    cv2.imwrite(save_name, original_img, (cv2.IMWRITE_JPEG_QUALITY, quality))
    
    img_jpg = cv2.imread(save_name)

    plt.figure(figsize=(20,10))
    plt.imshow(cv2.cvtColor(img_jpg, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
    
    return original_img, img_jpg, save_name

    
def print_stats(original_img, img_jpg, img_loc, img_jpg_loc):
    
    value_CR = calculateCR(img_loc, img_jpg_loc)
    value_SNR = calculateSNR(original_img, img_jpg)
    value_PSNR = calculatePSNR(original_img, img_jpg)

    jpg_size = os.stat(img_jpg_loc).st_size

    display(HTML('<p>'\
                'Image Array Type: '+str(img_jpg.dtype)+'<br>'\
                'Image Dimensions: '+str(img_jpg.shape[0])+'x'+str(img_jpg.shape[1])+'<br>'\
                'Channels: RGB'+str(img_jpg.shape[2])+'<br>'\
                'Compression Rate: '+str(round(value_CR, 2))+'<br>'\
                'SNR: '+str(round(value_SNR, 2))+' dB<br>'\
                'PSNR: '+str(round(value_PSNR, 2))+' dB<br>'\
                'Compressed File Size: '+str(jpg_size/1000.)+' KB</p></p><hr>'))
    
def show(img_loc, quality):
    
    original_img, img_jpg, img_jpg_loc = compress_to_jpg(img_loc, quality)
    print_stats(original_img, img_jpg, img_loc, img_jpg_loc)    


interact(show, img_loc = fixed("lenac.tif"), \
         quality = widgets.IntSlider(min=0,max=100,step=5,value=0))

display(HTML('<hr><br><br>'))

interact(show, img_loc = fixed("lena.tiff"), \
         quality = widgets.IntSlider(min=0,max=100,step=5,value=0))
```

<h5>7. Crie uma fun√ß√£o que apresente uma imagem (100√ó100) como se apresenta na Ô¨Ågura. O √¢ngulo de cada sector √© dado por par√¢metro passado √† fun√ß√£o (o √¢ngulo √© um valor inteiro entre 0 e 360 graus).</h5>
<hr>


```python
def radial_lines_img(angle, size):
    
    x = np.arange(-size, size, 1)
    y = np.arange(-size, size, 1)*1j
    x, y = np.meshgrid(x,y)
    z = x + y
    np.set_printoptions(threshold=100)
    lines = np.floor(np.angle(z)*angle).astype(np.int32)%2 * 255

    plt.figure(figsize=(20,10))
    plt.imshow(lines, cmap = 'gray')
    plt.axis('off')
    plt.show()

interact(radial_lines_img, angle = widgets.IntSlider(min=0,max=360,step=2,value=60), \
         size = fixed(500))
```

<h5>8. Construa uma fun√ß√£o que realize o algoritmo de dithering Floyd Steinberg. Esta fun√ß√£o recebe uma matrix (com os pixeis em tons de cinzento) e devolve uma matrix com valores a preto e branco. Este algoritmo aproxima cada pixel da imagem (x) ao valor mais pr√≥ximo (preto ou branco) e o erro √© difundido para os pixeis adjacentes seguindo o m√©todo: </h5>
<hr>


```python
def dither_one_channel(image, qfactor):
    
    iwidth = image.shape[0]
    iheight = image.shape[1]
    
    for x in range(0, iwidth-1):
        for y in range(0, iheight-1):

            pixel = image[x][y]

            qpixel = np.round(qfactor*pixel/255)*(255/qfactor)

            error = pixel - qpixel

            image[x    ][y    ] = qpixel
            image[x + 1][y    ] += ((7/16.) * error)
            image[x - 1][y + 1] += ((3/16.) * error)
            image[x    ][y + 1] += ((5/16.) * error)
            image[x - 1][y + 1] += ((1/16.) * error)

    return image
```

```python
def dither_multi_channel(image, qfactor):
    
    iwidth = image.shape[0]
    iheight = image.shape[1]
    ichannel = image.shape[2]
    
    for channel in range(0, ichannel):
        image[:,:,channel] = dither_one_channel(image[:,:,channel], qfactor)
        
    return image
```

```python
factor = 16

def show_dither_color(factor):
    
    color_img, color_size = get_img("lenac.tif")

    original_img = cv2.imread("lenac.tif")
    dithered_image = dither_one_channel(original_img.astype(np.float64), factor)
    cv2.imwrite("ex8_dither/lena_dither_color.jpg", dithered_image, (cv2.IMWRITE_JPEG_QUALITY, 100))
    dither_color_size = os.path.getsize("ex8_dither/lena_dither_color.jpg")

    plt.figure(figsize=(20,20))
    plt.title("Dithered Image :: RGB Channels :: File size " + str(dither_color_size/1000) \
                            + " Kbytes :: " + str(factor) + " colors", fontsize=20)
    plt.imshow(cv2.cvtColor(dithered_image.astype(np.uint8), cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    plt.figure(figsize=(20,20))
    plt.subplot(231)
    plt.title("Dithered Image :: R Channel :: " + str(factor) + " colors")
    plt.imshow(dithered_image[:,:,0], cmap='Reds_r')
    plt.axis('off')
    plt.subplot(232)
    plt.title("Dithered Image :: G Channel :: " + str(factor) + " colors")
    plt.imshow(dithered_image[:,:,1],  cmap='Greens_r')
    plt.axis('off')
    plt.subplot(233)
    plt.title("Dithered Image :: B Channel :: " + str(factor) + " colors")
    plt.imshow(dithered_image[:,:,2], cmap='Blues_r')
    plt.axis('off')
    plt.show()

    print("Original Color Size: " + str(color_size))
    print("Dithered Color Size: " + str(dither_color_size))
    
    
def show_dither_gray(factor):   
    
    color_img, color_size = get_img("lenac.tif")
    gray_img, gray_size = get_img("lena.tiff")

    gray_img = cv2.cvtColor(color_img, cv2.COLOR_BGR2GRAY)
    save_name = "ex8_dither/lena_bw.bmp"
    cv2.imwrite(save_name, gray_img)

    img_bw = cv2.imread(save_name)
    dithered_image = dither_one_channel(img_bw.astype(np.float64), factor)
    cv2.imwrite("ex8_dither/lena_dither_gray.jpg", dithered_image, (cv2.IMWRITE_JPEG_QUALITY, 100))
    dither_gray_size = os.path.getsize("ex8_dither/lena_dither_gray.jpg")

    plt.figure(figsize=(20,20))
    plt.title("Dithered Image :: GRAYSCALE :: File size " + str(dither_gray_size/1000) \
                            + " Kbytes :: " + str(factor) + " colors", fontsize=20)
    plt.imshow(dithered_image[:,:,0], cmap='gray')
    plt.axis('off')
    plt.show()

    plt.figure(figsize=(20,20))
    plt.subplot(231)
    plt.title("Dithered Image :: R Channel :: GRAYSCALE :: " + str(factor) + " colors")
    plt.imshow(dithered_image[:,:,0], cmap='gray')
    plt.axis('off')
    plt.subplot(232)
    plt.title("Dithered Image :: G Channel :: GRAYSCALE :: " + str(factor) + " colors")
    plt.imshow(dithered_image[:,:,1], cmap='gray')
    plt.axis('off')
    plt.subplot(233)
    plt.title("Dithered Image :: B Channel :: GRAYSCALE :: " + str(factor) + " colors")
    plt.imshow(dithered_image[:,:,2], cmap='gray')
    plt.axis('off')
    plt.show()
    
    print("Original Gray Size:  " + str(gray_size))
    print("Dithered Gray Size:  " + str(dither_gray_size))
    
color_img, color_size = get_img("lenac.tif")
plt.figure(figsize=(20,20))
plt.title("Original Image :: RGB (Inverted Channels from BGR) :: File size " \
          + str(color_size/1000) + " Kbytes", fontsize=20)
plt.imshow(cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB)) #Inverting from BGR Channels
plt.axis('off')
plt.show()
    
interact(show_dither_gray, factor = widgets.IntSlider(min=1,max=200,step=1,value=1))
interact(show_dither_color, factor = widgets.IntSlider(min=1,max=200,step=1,value=1))
```
