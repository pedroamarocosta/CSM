<h5>2. Grave a mesma imagem, mas agora em formato "JPEG" com diferentes qualidades. Verifique visualmente a qualidade das imagens assim como o tamanho do ficheiro. Calcule a taxa de compress√£o, a SNR e a PSNR.</h5>
<hr>

```python
def calculateCR(originalFile, compressedFile):

    return os.stat(originalFile).st_size/os.stat(compressedFile).st_size

def calculateSNR(originalImage, compressedImage):
    
    signal = np.sum(np.power(compressedImage, 2))
    error = np.sum(np.power((originalImage - compressedImage), 2))
    return 10*np.log10(signal/error)


def calculatePSNR(originalImage, compressedImage):
    
    maxValue = np.power(np.max(originalImage), 2)
    meanSquaredError = (1/(3*compressedImage.shape[0]*compressedImage.shape[1])) * \
                        np.sum(np.power((originalImage-compressedImage), 2))
    return 10 * np.log10(maxValue/meanSquaredError)


def compressToJPG(originalImagePath, quality):
    
    compressedImagePath = 'ex2/compressedImage(quality='+ str(quality) + ').jpg'
    originalImage = cv2.imread(originalImagePath)
    cv2.imwrite(compressedImagePath, originalImage, (cv2.IMWRITE_JPEG_QUALITY, quality))
    compressedImage = cv2.imread(compressedImagePath)
    plt.figure(figsize=(10,10))
    plt.imshow(cv2.cvtColor(compressedImage, cv2.COLOR_BGR2RGB))
    plt.show()
    return originalImage, compressedImage, compressedImagePath

    
def printImageData(originalImage, compressedImage, originalImagePath, compressedImagePath):
    
    valueCR = calculateCR(originalImagePath, compressedImagePath)
    valueSNR = calculateSNR(originalImage, compressedImage)
    valuePSNR = calculatePSNR(originalImage, compressedImage)
    compressedImageSize = os.stat(compressedImagePath).st_size
    display(HTML('<div style="font-size:16px; "Arial Black", Gadget, sans-serif;"><hr>'\
                'File: ' + str(compressedImagePath) + '<br>' \
                'Image Array Type: ' + str(originalImage.dtype) + '<br>'\
                'Image Dimensions: ' + str(originalImage.shape[0]) + 'x' + str(originalImage.shape[1]) + '<br>'\
                'Channels: RGB'      + str(originalImage.shape[2]) + '<br>'\
                'Compression Rate: ' + str(round(valueCR, 2)) + '<br>'\
                'SNR: ' + str(round(valueSNR, 2)) +' dB<br>'\
                'PSNR: ' + str(round(valuePSNR, 2)) +' dB<br>'\
                'Compressed File Size: ' + str(compressedImageSize/1000.) + ' KB</div><hr>'))
    
def show(originalImagePath, quality):
    
    originalImage, compressedImage, compressedImagePath = compressToJPG(originalImagePath, quality)
    printImageData(originalImage, compressedImage, originalImagePath, compressedImagePath)    

    
interact(show, originalImagePath = fixed("lenac.tif"), \
         quality = widgets.IntSlider(min=0,max=100,step=5,value=0))

display(HTML('<br><br>'))

interact(show, originalImagePath = fixed("lena.tiff"), \
         quality = widgets.IntSlider(min=0,max=100,step=5,value=0))
```

<h5>7. Crie uma fun√ß√£o que apresente uma imagem (100√ó100) como se apresenta na Ô¨Ågura. O √¢ngulo de cada sector √© dado por par√¢metro passado √† fun√ß√£o (o √¢ngulo √© um valor inteiro entre 0 e 360 graus).</h5>
<hr>


```python
def radialLinesImage(angle, size):
    
    x = np.arange(-size, size, 1)
    y = np.arange(-size, size, 1)*1j
    x, y = np.meshgrid(x,y)
    z = x + y
    
    lines = np.floor(np.angle(z)*angle).astype(np.int32)%2 * 255

    plt.figure(figsize=(20,10))
    plt.imshow(lines, cmap = 'gray')
    plt.show()

interact(radialLinesImage, angle = widgets.IntSlider(min=0,max=360,step=2,value=60), \
         size = fixed(500))
```

<h5>8. Construa uma fun√ß√£o que realize o algoritmo de dithering Floyd Steinberg. Esta fun√ß√£o recebe uma matrix (com os pixeis em tons de cinzento) e devolve uma matrix com valores a preto e branco. Este algoritmo aproxima cada pixel da imagem (x) ao valor mais pr√≥ximo (preto ou branco) e o erro √© difundido para os pixeis adjacentes seguindo o m√©todo: </h5>
<hr>


```python
def ditherOneChannel(image, qfactor):
    
    image = image.astype(np.float32)
    iwidth = image.shape[0]
    iheight = image.shape[1]
    
    for x in range(0, iwidth-1):
        for y in range(0, iheight-1):

            pixel = image[x][y]

            qpixel = np.round(qfactor*pixel/255)*(255/qfactor)

            error = pixel - qpixel

            image[x    ][y    ] = qpixel
            image[x + 1][y    ] += ((7/16.) * error)
            image[x - 1][y + 1] += ((3/16.) * error)
            image[x    ][y + 1] += ((5/16.) * error)
            image[x - 1][y + 1] += ((1/16.) * error)

    return image
```

```python
def ditherMultiChannel(image, qfactor):
    
    iwidth = image.shape[0]
    iheight = image.shape[1]
    ichannel = image.shape[2]
    
    for channel in range(0, ichannel):
        image[:,:,channel] = ditherOneChannel(image[:,:,channel], qfactor)
        
    return image
```

<h5>9. Construa uma fun√ß√£o para gravar a matriz para um Ô¨Åcheiro bin√°rio. VeriÔ¨Åque o tamanho do Ô¨Åcheiro inicial e do Ô¨Åcheiro Ô¨Ånal. Calcule a taxa de compress√£o e me√ßa o SNR e o PSNR.</h5>
<hr>


```python
def writeBinary(array, filename):
    
    file = open(filename, 'w')
    
    width = array.shape[0]
    height = array.shape[1]
    channel = array.shape[2]
    
    for i in range(width):
        for j in range(height):
            for c in range(channel):
                file.write(np.binary_repr(array[i][j][c]))
    
    return file

originalImagePath = "lena.tiff"
binaryFile = "ex9/lenagraydither.txt"
ditheredGrayBinaryFile = writeBinary(ditheredGrayImage, binaryFile)

originalImagePath = "lenac.tif"
binaryFile = "ex9/lenacolordither.txt"
ditheredColorBinaryFile = writeBinary(ditheredColorImage, binaryFile)
```
